```python
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModelForSequenceClassification, get_linear_schedule_with_warmup
from torch.optim import AdamW
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Dataset class ---
class NewsDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len=256):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = self.tokenizer(
            self.texts[idx],
            truncation=True,
            padding='max_length',
            max_length=self.max_len,
            return_tensors='pt'
        )
        item = {key: val.squeeze(0) for key, val in encoding.items()}
        item['labels'] = torch.tensor(self.labels[idx], dtype=torch.long)
        return item

# --- 2. Load data ---
df = pd.read_csv('/content/complete_dataset.csv')
df['text'] = df['text'].astype(str).fillna('')
train_texts, val_texts, train_labels, val_labels = train_test_split(
    df['text'].tolist(), df['label'].tolist(), test_size=0.8, random_state=42, stratify=df['label']
)

# --- 3. Load MuRIL tokenizer and model ---
MODEL_NAME = "google/muril-base-cased"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=2)

# --- 4. DataLoaders ---
train_dataset = NewsDataset(train_texts, train_labels, tokenizer)
val_dataset = NewsDataset(val_texts, val_labels, tokenizer)
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

# --- 5. Training setup ---
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)
optimizer = AdamW(model.parameters(), lr=2e-5)
num_epochs = 10
total_steps = len(train_loader) * num_epochs
scheduler = get_linear_schedule_with_warmup(
    optimizer, num_warmup_steps=0, num_training_steps=total_steps
)

# --- 6. Training loop with tracking ---
train_losses = []
val_accuracies = []

for epoch in range(num_epochs):
    model.train()
    train_loss = 0
    for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}"):
        optimizer.zero_grad()
        batch = {k: v.to(device) for k, v in batch.items()}
        outputs = model(**batch)
        loss = outputs.loss
        loss.backward()
        optimizer.step()
        scheduler.step()
        train_loss += loss.item()
    avg_train_loss = train_loss / len(train_loader)
    train_losses.append(avg_train_loss)
    print(f"Epoch {epoch+1} - Train Loss: {avg_train_loss:.4f}")

    # Validation
    model.eval()
    preds, true_labels = [], []
    with torch.no_grad():
        for batch in val_loader:
            batch = {k: v.to(device) for k, v in batch.items()}
            outputs = model(**batch)
            logits = outputs.logits
            preds.extend(torch.argmax(logits, dim=1).cpu().numpy())
            true_labels.extend(batch['labels'].cpu().numpy())
    acc = accuracy_score(true_labels, preds)
    val_accuracies.append(acc)
    print(f"Epoch {epoch+1} - Validation Accuracy: {acc:.4f}")

# --- 7. Save the trained model and tokenizer ---
save_path = "/content/muril_fakenews_model"
model.save_pretrained(save_path)
tokenizer.save_pretrained(save_path)
print(f"Model and tokenizer saved at {save_path}")

# --- 9. Final Evaluation ---
print("\nClassification Report (Validation):")
print(classification_report(true_labels, preds, target_names=['REAL', 'FAKE']))

# --- 10. Confusion Matrix ---
cm = confusion_matrix(true_labels, preds)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['REAL', 'FAKE'], yticklabels=['REAL', 'FAKE'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix (Validation)')
plt.show()
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification # Removed GenerationMixin import
from newspaper import Article
import trafilatura
from readability import Document
import requests

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
# Ensure the model is loaded before moving it to device
# If model is not already loaded in a previous cell, load it here
# MODEL_NAME = "google/muril-base-cased"
# model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=2)
model.to(device)

MODEL_NAME = "google/muril-base-cased"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

# 2. Prediction function
def predict_news(text, threshold=0.5):
    global tokenizer
    inputs = tokenizer(
        text,
        truncation=True,
        padding='max_length',
        max_length=512,
        return_tensors='pt'
    )
    inputs = {k: v.to(device) for k, v in inputs.items()}
    with torch.no_grad():
        outputs = model(**inputs)
        probs = torch.softmax(outputs.logits, dim=1).cpu().numpy()[0]
        fake_prob = probs[0]
        real_prob = probs[1]
        if real_prob > threshold:
            label = "REAL"
            confidence = real_prob
        else:
            label = "FAKE"
            confidence = fake_prob
    print(f"Prediction: {label} (Confidence: {confidence*100:.2f}%)")
    return label, confidence

# 3. Combined extraction function
def get_news_text_from_url(url):
    # Try newspaper3k
    try:
        article = Article(url)
        article.download()
        article.parse()
        if article.text and len(article.text.strip()) > 100:
            print("Extracted using newspaper3k.")
            return article.text
    except Exception as e:
        pass

    # Try trafilatura
    try:
        downloaded = trafilatura.fetch_url(url)
        if downloaded:
            text = trafilatura.extract(downloaded)
            if text and len(text.strip()) > 100:
                print("Extracted using trafilatura.")
                return text
    except Exception as e:
        pass

    # Try readability-lxml
    try:
        response = requests.get(url, timeout=10)
        if response.status_code == 200:
            doc = Document(response.text)
            content = doc.summary()
            # Remove HTML tags
            import re
            cleanr = re.compile('<.*?>')
            cleantext = re.sub(cleanr, '', content)
            if cleantext and len(cleantext.strip()) > 100:
                print("Extracted using readability-lxml.")
                return cleantext
    except Exception as e:
        pass
    print("Failed to extract clean text from the URL.")
    return None

# 4. Example Usage

# Test with a headline
headline = "India of today not India of 1975: Shashi Tharoor criticises Emergency in op-ed"
predict_news(headline)

# Test with a news article from a URL
url = "https://www.indiatoday.in/india/story/shashi-tharoor-column-sharp-criticism-emergency-anniversary-indira-gandhi-2753556-2025-07-10"
news_text = get_news_text_from_url(url)
if news_text and news_text.strip():
    predict_news(news_text)
else:
    print("Prediction not performed due to extraction failure.")
```
